```python
# Read the four files we deownloaded
import pandas as pd

prod_df = pd.read_csv('datasets/prod_dimen.csv') 

orders_df = pd.read_csv('datasets/orders_dimen.csv')

market_df = pd.read_excel('/datasets/market_fact.xlsx')

cust_df = pd.read_excel('/datasets/cust_dimen.xlsx')
```

### Let us start merging (***Joining***) the dataframes
- `pd.merge()`

```python
# Browse files one by one
# product file
cust_df.tail()
```

```python
# Products Dataframe
market_df.sort_values(by='Cust_id', ascending=True).head()
```

```python
# for merging them depending on `market_df`
first_merge_df = pd.merge(market_df, cust_df, how='inner', on='Cust_id')   # `inner` is the type of join, `on` is the equal values in both sides
first_merge_df.head()
```

```python
# Lets join the two dataframe depending on `prod_df
second_merge_df = pd.merge(prod_df, market_df, how='outer', on='Prod_id')
second_merge_df.sort_values(by='Prod_id', ascending=True)
second_merge_df.head(10)
```

```python
# Lets sort the dataframe depends on the order_id to see the values
second_merge_df.sort_values(by='Ord_id', ascending=True).head()
```
